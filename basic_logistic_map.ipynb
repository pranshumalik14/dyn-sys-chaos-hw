# An Initial Exploration of the Logistic Map

$$x_{n+1} = ax_n(1-x_n)$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logistic_map(x, a=1):
    return a*x*(1-x)

def logistic_series(x0, n, a=1):
    xs = [x0]
    for _ in range(n):
        xi = logistic_map(xs[-1], a)
        xs.append(xi)
    return xs
```

```python
xs = logistic_series(0.1, 15, a=2)
plt.plot(xs, marker='o')
plt.title('Logistic Map Series (a=2, x0=0.1)')
```

```python
def tail_values(x0, n_total, n_skip, a):
    xs = logistic_series(x0, n_total, a=a)
    return xs[n_skip:]
```

### Varying $a$

```python
def long_run_values(a_values, x0, n_total=300, n_skip=200):
    xs_all = []
    as_all = []
    for a in a_values:
        tail = tail_values(x0, n_total, n_skip, a)
        xs_all.extend(tail)
        as_all.extend([a] * len(tail))
    return as_all, xs_all
```

```python
avals = np.linspace(0.01, 4.0, 500)
as_all, xs_all = long_run_values(avals, x0=0.1)
plt.plot(as_all, xs_all, '.', markersize=1, alpha=0.5)
plt.vlines(3.56994, 0, 1, color='k', lw=1)
plt.annotate('Chaos Boundary', xy=(2.5, 0.1))
plt.xlabel('a')
plt.ylabel('x over 100 steps (after discarding first 200 steps)')
plt.title('Logistic Map: Long-Run Values (x0 = 0.1)')
```

```python
avals = np.linspace(-0.01, -2, 250)
as_all, xs_all = long_run_values(avals, x0=0.1)
plt.plot(as_all, xs_all, '.', markersize=1, alpha=0.5)
plt.xlabel('a')
plt.ylabel('x over 100 steps (after discarding first 200 steps)')
plt.title('Logistic Map: Long-Run Values (x0 = 0.1)')
```

Series seems to diverge after $a = 4$.

```python
avals = np.linspace(4.01, 8.0, 500)
as_all, xs_all = long_run_values(avals, x0=0.1, n_total=3, n_skip=0)
plt.plot(as_all, xs_all, '.', markersize=1, alpha=0.5)
plt.xlabel('a')
plt.ylabel('x over first 3 steps')
plt.title('Logistic Map: Long-Run Values (x0 = 0.1)')
```

```python
avals = np.linspace(4.01, 8.0, 500)
as_all, xs_all = long_run_values(avals, x0=0.1, n_total=4, n_skip=0)
plt.plot(as_all, xs_all, '.', markersize=1, alpha=0.5)
plt.xlabel('a')
plt.ylabel('x over first 4 steps')
plt.title('Logistic Map: Long-Run Values (x0 = 0.1)')
```

### Varying $x_0$

```python
avals = np.linspace(0.01, 4.0, 500)
as_all, xs_all = long_run_values(avals, x0=0.101)
plt.plot(as_all, xs_all, '.', markersize=1, alpha=0.5)
plt.vlines(3.56994, 0, 1, color='k', lw=1)
plt.annotate('Chaos Boundary', xy=(2.5, 0.1))
plt.xlabel('a')
plt.ylabel('x over 100 steps (after discarding first 200 steps)')
plt.title('Logistic Map: Long-Run Values (x0 = 0.101)')
```

Nothing much seems to change with tiny deviations from $x_0 = 0.1$, compared to plots in the previous section. For $x_0 <0$ or $x_0 > 1$, we will have $x_{1:n} < 0$ which will continually decrease over increasing $n$.

```python
avals = np.linspace(0, 4.0, 500)
as_all, xs_all = long_run_values(avals, x0=-0.5, n_total=2, n_skip=0)
plt.plot(as_all, xs_all, '.', markersize=1, alpha=0.5)
plt.xlabel('a')
plt.ylabel('x over first 2 steps')
plt.title('Logistic Map: Long-Run Values (x0 = -0.5)')
```
